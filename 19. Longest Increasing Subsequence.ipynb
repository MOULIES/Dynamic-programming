```python
def lis( arr ):
    n = len( arr )
    dp= [1]*n
    
    for i in range( 1, n):
        for j in range( i ):
            if arr[i] > arr[j] and dp[i] < dp[j]+1 :
                dp[i] = dp[j]+1
    return max( dp )
```

```python
a = [ 5,1,8,10,5,2,14,17,19,7,9,29]
lis(a)
# ans : 1,8,10,14,17,19,29
```

```python
def lis1( arr ):
    n = len( arr )
    dp= [1]*n
    
    for i in range( n-1, -1 , -1):
        for j in range( i+1 , n ):
            if arr[i] < arr[j] :
                dp[i] = max( dp[i] , 1 + dp[j] )
    return max( dp )
```

```python
arr = [10, 22, 9, 33, 21, 50, 41, 60]
lis1(arr)
```

```python
def lis3( arr ):
    n = len( arr )
    dp= [ 1 for i in range(n) ]
    
    for i in range( 1, n):
        for j in range( i-1 , -1, -1 ):
            if arr[j] < arr[i]  :
                dp[i] = max( dp[i] , dp[j] + 1 )
    return max(dp), dp
a = [ 5,1,8,10,5,2,14,17,19,7,9,29]
lis3(a)
```

### binary search

```python
def binarySearch1( arr , x ):
    l = 0
    h = len( arr)-1
    import math
    ans = math.inf
    
    while l <= h :
        mid = l+(h-l )//2
        
        if arr[mid] >= x :
            ans = min( ans, mid)
            h = mid -1
        else:
            l = mid+1
    return ans if ans != math.inf else -1
            
    

def lengthOfLIS1( nums):
    dp = []
    dp.append( nums[0] )
    n = len(nums)
    for i in range( 1, n ):
        a1 = binarySearch1( dp , nums[i] )
        if a1 == -1:
            dp.append( nums[i] )
        else:
            dp[a1] = nums[i]
#         print( dp )
    return len(dp ) , dp

a = [ 1,5 ,8 ,10,5,2,14,17,19,7,9,29,28]
print(lengthOfLIS1(a))
```

```python

def bisect_left(a, x, lo=0, hi=None):

    if lo < 0:
        raise ValueError('lo must be non-negative')
    if hi is None:
        hi = len(a)
    while lo < hi:
        mid = (lo+hi)//2
        # Use __lt__ to match the logic in list.sort() and in heapq
        if a[mid] < x: lo = mid+1
        else: hi = mid
    return lo if lo < len(a) else -1


def lengthOfLIS3( nums):
    dp = []
    dp.append( nums[0] )
    n = len(nums)
    for i in range( 1, n ):
        a1 = bisect_left( dp , nums[i] )
        if a1 == -1:
            dp.append( nums[i] )
        else:
            dp[a1] = nums[i]
#         print( dp )
    return len(dp ) , dp

a = [ 1,5 ,8 ,10,5,2,14,17,19,7,9,29,28]
print(lengthOfLIS3(a))
```
