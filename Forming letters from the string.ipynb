```python
from itertools import product

dp = {}
k = 65
for i in range( 1 , 27 ):
    dp[i] = chr(k)
    k += 1
    
s = input()
vowels , total = 0 , 0
length = len( s )
for i in s :
    if ord(i)-64 in ( 1,5,9,15,21):
        vowels += 1
    total += ord(i) - 64 


arr = product( dp.keys() , repeat = length ) 

new_arr = []
for i in arr :
    if sum( i ) == total:
        vow_c = 0
        word = ''
        for vow in i :
            word += dp[vow]
            if vow in ( 1,5,9,15,21):
                vow_c += 1
        if vow_c == vowels:
            new_arr.append( word )
            
new_arr
```

```python
def permute_string(str):
    if len(str) == 0:
        return ['']
    prev_list = permute_string(str[1:len(str)])
    # print(prev_list,"rev")
    next_list = []
    for i in range(0,len(prev_list)):
        for j in range(0,len(str)):
            new_str = prev_list[i][0:j]+str[0]+prev_list[i][j:len(str)-1]
#             if new_str not in next_list:
            next_list.append(new_str)
    next_list.sort()
    return next_list

print(permute_string('AID'));
```
