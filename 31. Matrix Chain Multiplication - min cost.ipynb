```python
def MCM(arr , l , r):
    if l>= r : return 0
    
    ans = float('inf')
    
    for k in range(l,r):
        temp = ( MCM( arr, l, k) + #left
                MCM( arr , k+1 , r ) + 
                arr[l-1]*arr[k]*arr[r]   )
        ans = min( temp, ans)
    return ans
    
```

```python
arr = [1, 2, 3, 4, 3]
n = len(arr)
 
print("Minimum number of multiplications is ",MCM(arr, 1, n-1))

arr = [40, 20, 30, 10 ,30]
n = len(arr)
 
print("Minimum number of multiplications is ", MCM(arr, 1, n-1))
```

```python
def MCM_memoized(arr , n):
    mem = {}
    def dfs(arr , l , r):
        if l>= r : return 0
        if (l,r) in mem: return mem[(l,r)]
        ans = float('inf')

        for k in range(l,r):
            temp = ( MCM( arr, l, k) + #left
                    MCM( arr , k+1 , r ) +  #right
                    arr[l-1]*arr[k]*arr[r]   )  #cost
            ans = min( temp, ans)
        mem[(l,r) ] = ans
        return  mem[(l,r)]
    return dfs( arr , 1 , n-1)
```

```python
arr = [1, 2, 3, 4, 3]
n = len(arr)
print("Minimum number of multiplications is ",MCM_memoized(arr,  n ))

arr = [40, 20, 30, 10 ,30]
n = len(arr)
print("Minimum number of multiplications is ", MCM_memoized(arr, n ))
```

```python
def mcm_dp( arr , n ):
    dp = [ [ 0 ]*n for _ in range( n ) ]
    
    for g in range(2,n):
        for l in range(1, n-g+1):
            r = l+g-1
            dp[l][r] = float('inf')
            for k in range(l, r):
                tc = dp[l][k] + dp[k+1][r] + ( arr[l-1]* arr[k]*arr[r] )
                dp[l][r] = min( dp[l][r] ,tc )
#     for i in dp:
#         print(i)
    return dp[1][n-1]
```

```python
arr = [1, 2, 3, 4,3]
n = len(arr) 
print("Minimum number of multiplications is ",mcm_dp(arr,  n ))

arr = [40, 20, 30, 10 ,30]
n = len(arr)
print("Minimum number of multiplications is ", mcm_dp(arr, n ))
```
