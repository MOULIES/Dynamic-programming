```python

def isSubsetSum( arr , n , summ):
    
    dp= [ [False]*(summ+1) for _ in range(n+1)]
    
    for i in range( n+1):
        dp[i][0] = True
        
    for i in range( 1, n+1):
        for j in range( 1 , summ+1):
            if j < arr[i-1]:
                dp[i][j] = dp[i][j-1]
            else:
                dp[i][j] = dp[i][j-1] or dp[i][j- arr[i-1] ]
    
    return dp[-1]
```

```python
def minSubsetDiff( arr , n ): 
    s = sum( arr)
                                               
    #last row of subset sum dp
    dp = isSubsetSum( arr , n , s//2 )  #here also  we can can pass as s//2 .. it works
                                               
    import math
    diff = math.inf
    
    for i in range( s//2 +1 ):
        s1 = i
        s2 = s - i
        
        if dp[i] == True and diff > abs(s1 -s2):
            diff = abs( s1 - s2)
            
    return diff
                                               
                                        
```

```python
arr = [1,6,11,5]
minSubsetDiff( arr , len( arr) )
```
