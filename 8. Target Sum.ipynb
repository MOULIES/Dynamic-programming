```python
def  countSubsetSum( arr ,  n,  total):
#     base cases
    if total == 0 : return 1
    
    if n==0 and total != 0 : return 0
    
    if total < arr[n-1]:
        return countSubsetSum( arr ,  n-1 ,  total )
    else:
        return countSubsetSum( arr ,  n-1 ,  total ) +  countSubsetSum( arr ,  n-1 ,  total - arr[n-1] )
```

```python
def countSubsetSum_dp( arr , n, summ ):
    dp= [ [0]*(summ+1) for _ in range( n+1)]
    
    for i in range(n+1):
        dp[i][0] = 1
    
    for i in range( 1, n +1 ):
        for j in range( 1, summ+1):
            if j < arr[i-1]:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j] + dp[i-1][j - arr[i-1]]
    return dp[-1][-1]
```

```python
def tagetSum( arr , target):
    n = len(arr)
    sum_ = sum( arr )
    
    s1 = (sum_ + target)//2
    
    if s1 > ( sum_ - s1 ):
        s1 = sum_-s1
        
    return countSubsetSum_dp( arr , n, s1)
```

```python
arr = [1,1,1,1,1]
target = 3
tagetSum( arr , target)
```
