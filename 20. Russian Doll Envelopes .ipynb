```python
def maxEnvelopes( envelopes ):
        n  = len( envelopes )
    
        if n==0: return 0

        envelopes.sort( )

        lis = [1]*n
        
        for i in range( 1, n):
            for j in range(i):
                if envelopes[j][1] < envelopes[i][1] and envelopes[j][0] < envelopes[i][0] and lis[i] < 1+lis[j] :
                    lis[i] = 1+lis[j]
                    
                
        return max(lis)
    
#w,h
a = [[5,4], [6,4],[6,7],[2,3]]  #op- 3
print( maxEnvelopes(a) )
```

```python
def bisect_left( arr , x):
    lo = 0
    hi = len(arr) 
    while lo < hi :
        mid = (lo+hi)//2
        if arr[mid] < x: lo = mid +1
        else:
            hi = mid
    return lo 

def maxEnvelopes2( A ):
    #wdith asc & height desc
    A.sort(key=lambda x: ( x[0], -x[1] ) )
    lis = []
    size = 0
    for w in map(lambda x: x[1] , A ) :
        idx =  bisect_left( lis, w )
        i = idx if idx < size else -1
        if i == -1:
            lis.append( w )
            size += 1
        else:
            lis[i] = w
    return size
        
#w,h
A =[[4,5],[4,6],[6,7],[2,3],[1,1]]
# A = [[5,4],[6,4],[6,7],[2,3]]

#op 3
print( maxEnvelopes2(A) )
```
