```python
def largestRectangleArea( heights ):
    n = len( heights )
    left, right = [], []
    
    stack = []
    for i in range(n):
        if not stack :
            left.append( 0 )
            stack.append( i )
        else:
            while  stack and heights[ stack[-1] ] >= heights[i] :
                stack.pop()
            if not stack :
                left.append( 0 )
            else:
                left.append( stack[-1]+1 )
            
            stack.append(i)
    
    stack.clear()
    
    for i in range(n-1,-1,-1):
        if not stack:
            right.insert(0, n-1 )
            stack.append( i )
        else:
            while  stack and heights[ stack[-1] ] >= heights[i] :
                stack.pop()
                
            if not stack :
                right.insert( 0, n-1 )
            else:
                right.insert( 0 , stack[-1]-1 )
            stack.append(i)
        
    
    mx_area = 0
    for i in range(n):
        mx_area = max( mx_area , heights[i] * (right[i]-left[i]+1) )
        
    print(left ,'\n', right )
    return mx_area
```

```python
heights= [2,1,5,6,2,3]
print( largestRectangleArea( heights ) )
```

```python
def maxHist( row ) :
    maxArea = 0
    stack  = [ ]   #pair:( index, height )

    for i , h in enumerate( row ):
        start = i

        while stack and stack[-1][1] > h :
            index, height = stack.pop()
            maxArea = max( maxArea , height * ( i - index ))
            start = index
        stack.append( [start, h ] )

    for i , h in stack :
        maxArea = max( maxArea  , h*(len(row) -i)  )
    return maxArea
```

```python
heights= [2,1,5,6,2,3]

print(  maxHist( heights )  )
```
