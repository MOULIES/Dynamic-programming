```python
def minimumEditDistance( s1 , s2 ):
    m = len(s1)
    n = len(s2)
    
    dp = [ [ max(m,n) for _ in range( n+1)] for _ in range( m+1)]
    
    for i in range( m+1):
        for j in range( n+1 ):
            if j == 0:
                dp[i][j] = i
            elif i == 0:
                dp[i][j] = j
            elif  s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] ## character same and recur for remainind string
            else:
                dp[i][j] = 1 + min(  dp[i][j-1] , #insert
                                     dp[i-1][j] , #remove
                                     dp[i-1][j-1] ) #replace
    return dp[-1][-1]
```

```python
s1=  'sunday'
s2 = 'saturday'
minimumEditDistance( s1 , s2 )
```
