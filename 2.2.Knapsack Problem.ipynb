## Algorithm

```python
#recursive approach
# https://www.geeksforgeeks.org/python-program-for-dynamic-programming-set-10-0-1-knapsack-problem/

def knapsack( wt, profit, weight , n):
    if n == 0 or weight == 0 :
        return 0
    
    if wt[n-1] > weight :
        return knapsack( wt, profit, weight , n-1 )
    else:
        return max(  knapsack( wt, profit, weight , n-1 )  , 
                   
                     profit[n-1] + knapsack( wt, profit, weight - wt[n-1] , n-1 )  )
        
```

```python
profit = [10, 40, 30, 50]
wt = [5, 4, 6, 3]
weight = 10
n = 4
knapsack( wt, profit, weight , n)
```

```python
#recursve with memoization

def knapsack1( wt, profit, weight , n ):
    mem = [ [ -1 for _ in range( weight+1)] for _ in range( len(wt)+1 ) ]
    
    def _knapsack( wt, profit, weight , n ):
        if n == 0 or weight == 0 :
            return 0
        
        if mem[n][weight] != -1 : return mem[n][weight]

        if wt[n-1] > weight :
            return _knapsack( wt, profit, weight , n-1 )
        else:
            mem[n][weight] =  max(  _knapsack( wt, profit, weight , n-1 )  , 

                         profit[n-1] + _knapsack( wt, profit, weight - wt[n-1] , n-1 )  )
            return mem[n][weight]

    _knapsack( wt, profit, weight , n )
    return mem[-1][-1]
```

```python
profit = [ 10, 40, 30, 50 ]
wt = [5, 4, 6, 3]
weight = 10
n = 4
knapsack1( wt, profit, weight , n )
```

```python
#bounded problem limit o <= x <= 1

def knapsack_dp( wt, profit , capacity , n):
    if n == 0 or capacity == 0:
        return 0
    dp = [ [  0 for _ in range( capacity +1 )] for _ in range( n + 1) ]
    
    for i in range( 1, n+1 ):
        for j in range(  1, capacity+1 ):
            if j < wt[i-1] :
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = max( dp[i-1][j]  , profit[i-1] +  dp[i-1][j - wt[i-1] ]  )

    return dp[-1][-1]
```
