```python
def countSubsetSum( arr , n, s ):
    dp = [ [0]*(s+1) for i in range( n+ 1) ]
    
    dp[0][0] = 1
    for i in range( 1, n+ 1):
        for j in range(  s+1):
            if j < arr[i-1]:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j] +dp[i-1][ j - arr[i-1]]
    return dp[-1][-1]
```

```python
def countSubset_given_diff( arr , diff):
    n = len( arr )
    
    s = sum(arr)
    
    new_sum = (s+diff)//2
    
    return countSubsetSum( arr , n,new_sum )
    
```

```python
s = [3,1,2,3]
diff = 3
countSubset_given_diff( s , diff)
```
