```python
def minPathSum( grid):
    m, n  = len( grid), len( grid[0] )
    
    dp = [ [0]*n for i in range(m)]
    
    dp[0][0] = grid[0][0]
    
    #1st colums sum
    for i in range(1,m):
        dp[i][0] = grid[i][0]+dp[i-1][0]
    
    #1st row sum
    for i in range(1,n):
        dp[0][i] =  grid[i][0]+ dp[0][i-1]
        
    for i in range(1,m):
        for j in range( 1, n):
            dp[i][j] = grid[i][j]+ min( dp[i-1][j], dp[i][j-1])
    return dp[-1][-1]
```

```python
g = [ [1,3,5] ,
    [2,1,2],
    [4,3,1]] 
print(minPathSum( g) )
```

```python
def minPathSum22( grid):
    m, n  = len( grid), len( grid[0] )
    
    dp = [ [0]*(n) for i in range(m)]

    i = 0
    while  i+grid[i][0] < m :
        i = i+grid[i][0]
        dp[ i][0] =1
        
    i = 0
    while  i+grid[0][i] < n :
        i = i+grid[0][i]
        dp[ 0][i] =1

    m = 0
    for i in range(1,m):
        for j in range( 1, n):
            if dp[i-1][j]:
                w =  i +grid[i-1][j]
                if w <= m :
                    dp[w-1][j] = 1
                
                
            if dp[i][j-1] :
                w = i +grid[i][j-1]
                if w < n :
                    dp[i][w-1] = 1
                    
    if dp[-1][-1] == 0 : return -1
#     else:
        
    return dp[-1][-1]
```

```python
R,C = 2,3
# h =[ [1,1],[1,1] , [1,1]]
h = [[1,1,3], [1,1,1],[1,1,1]]

for i in range( int(input() ) ):
    R,C=list( map( int , input().split() ) )
    h = []
    for i in range(R):
        h.append( list( map( int , input().split() ) )  )
        
    mem = {}
    def isSafe(x, y):
        return (x < R and y < C)

    def minJump( x, y):
        if (x,y) in mem : return mem[(x,y)]

        if (x == R - 1 and y == C - 1):
            return 0

        down = 10**9
        if (isSafe( x + h[x][y] , y )):
            down =  1+ minJump( x + h[x][y] , y) 

        right = 10**9
        if (isSafe( x,  y + h[x][y]  )):
            right =  1+ minJump( x,  y + h[x][y] ) 

        mem[(x,y)] = min([down, right])
        return mem[(x,y)]

    ans = minJump( 0, 0)
    print( -1  if ans ==  10**9 else ans )
```

```python
def intersectionOfArrays(nums1 , nums2):
    return list( set(nums1) & set(nums2) )

nums1 = [1,2,2,1]
nums2 = [2,2]
print(intersectionOfArrays(nums1 , nums2))

nums1 = [4,9,5]
nums2 = [9,4,9,8,4]
print(intersectionOfArrays(nums1 , nums2))
```

```python
def intersectionOfArrays(nums1 , nums2):
    temp = []
    for i in nums1:
        if i in nums2 and i not in temp :
            temp.append(i)
    return temp

nums1 = [1,2,2,1]
nums2 = [2,2]
print(intersectionOfArrays(nums1 , nums2))

nums1 = [4,9,5]
nums2 = [9,4,9,8,4]
print(intersectionOfArrays(nums1 , nums2))
```
