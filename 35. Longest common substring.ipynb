```python
def longestCommonSusbtring(s1,s2):
    dp = [  [0 for _ in range(len(s2)+1)] for _ in range(len(s1)+1 ) ]
    
    longest = 0
    for i in range(1, len(s1)+1 ):
        for j in range(1,len(s2)+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = 1+dp[i-1][j-1]
                longest = max( longest , dp[i][j] )
                end_index = i
    return longest , s1[ end_index -longest :end_index]
                
```

```python
s1 = "abcdgh"
s2 ="acdghr"
print( longestCommonSusbtring(s1,s2) )
```
