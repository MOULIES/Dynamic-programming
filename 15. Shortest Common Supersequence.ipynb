```python
def findLCS(s1 , s2 ):
    
    dp = [ ['' for _ in range(len(s2)+1 )]  for _ in range( len(s1)+1) ] 
    
    for i in range( 1, len(s1)+1 ):
        for j in range( 1, len(s2)+1 ):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1]+s1[i-1]
            else:
                dp[i][j] =  dp[i-1][j]  if len( dp[i-1][j] ) > len( dp[i][j-1] ) else dp[i][j-1]
    return dp[-1][-1]
```

```python
def shortestCommonSupersequence( s1 , s2):
    ans = ""
    lcs = findLCS(s1 , s2 )
    
    p1, p2 = 0,0
    
    for ch in lcs :
        while s1[p1] != ch :
            ans += s1[p1]
            p1 +=1 
        while s2[p2] != ch :
            ans += s2[p2]
            p2 += 1
        
        ans += ch
        p1 +=1 
        p2 += 1
    ans += s1[p1:] + s2[p2:]
    return ans
        
```

```python
s1 = 'AGGTABST'
s2 = "GXTXAYBOP"
shortestCommonSupersequence( s1 , s2)
```
