```python
#recursive approach

def  isSubsetSum( arr ,  n,  total):
#     base cases
    if total == 0 : return True
    
    if n==0 and total != 0 : return False
    
    if total < arr[n-1]:
        return isSubsetSum( arr ,  n-1 ,  total )
    else:
        return isSubsetSum( arr ,  n-1 ,  total ) or  isSubsetSum( arr ,  n-1 ,  total - arr[n-1] )
```

```python
set = [3, 34, 4, 12, 5, 2]
sum = 13
n = len(set)
if (isSubsetSum(set, n, sum) == True):
    print("Found a subset with given sum")
else:
    print("No subset with given sum")
```

```python
#dp

def isSubsetSum_dp(arr , n, summ):
     
    dp =[ [False]*(summ + 1) for i in range(n + 1) ]
     
    # If sum is 0, then answer is true
    for i in range(n + 1):
        dp[i][0] = True
             
    # Fill the subset table in bottom up manner
    for i in range(1, n + 1):
        for j in range(1, summ + 1):
            if j< arr[i-1]:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = ( dp[i-1][j] or  dp[i - 1][j-arr[i-1]] )
                
#     for i in dp:
#         print(i)
    return dp[-1][-1]
```

```python
set = [3, 34, 4, 12, 5, 2]
sum = 13
n = len(set)
if (isSubsetSum_dp(set, n, sum) == True):
    print("Found a subset with given sum")
else:
    print("No subset with given sum")
```
