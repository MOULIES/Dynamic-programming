```python
def lcs(s1 , s2 ):
    
    dp = [ [0 for _ in range(len(s2)+1 )]  for _ in range( len(s1)+1) ] 
    
    for i in range( 1, len(s1)+1 ):
        for j in range( 1, len(s2)+1 ):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = 1+dp[i-1][j-1]
            else:
                dp[i][j] = max( dp[i-1][j] , dp[i][j-1])
    return dp[-1][-1]

def pattern_matching( s1, s2):
    if lcs(s1,s2) == len(s1):
        print( True)
    else:
        print( False)
```

```python
s1 = 'aab'
s2 = "abbabb"
pattern_matching( s1, s2)
```
