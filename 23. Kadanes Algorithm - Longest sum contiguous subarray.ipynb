```python
def maxSubArraySum( a ):
    msf = float('-inf')  #int_min
    meh = 0
    
    for i in range( len(a) ):
        meh += a[i]
        
        if meh < a[i] :
            meh = a[i]
        
        if msf < meh :
            msf = meh
    return msf
```

```python
a = [-2,-3,4,-1,-2,1,5,-3]
maxSubArraySum( a )
```

```python
def maxSubArraySum1( a ):
    msf = float('-inf')  #int_min
    meh = 0
    start = 0
    end = 0
    beg = 0
    for i in range( len(a) ):
        meh += a[i]
        
        if meh < a[i] :
            meh = a[i]
            beg = i
        
        if msf < meh :
            msf = meh
            start = beg
            end = i
    return start, end 
```

```python
a = [-2,-3,4,-1,-2,1,5,-3]
s,e = maxSubArraySum1( a ) 
print(a[ s:e+1] )
```
