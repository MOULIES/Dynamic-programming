```python
def uniquePathWithObstacles( grid ):
    rows = len( grid )
    cols = len( grid[0] )
    
    dp= [ [0]*cols for i in range( rows) ]
    
    flag = False # fill 1st row
    for i in range(cols):
        if flag or grid[0][i] == 1 :
            flag = True
            dp[0][i] = 0
        else:
            dp[0][i] = 1
            
    flag = False #fill 1st col
    for i in range(rows):
        if flag or grid[i][0] == 1 :
            flag = True
            dp[i][0] = 0
        else:
            dp[i][0] = 1
            
    for i in range(1,rows):
        for j in range(1,cols):
            if grid[i][j] == 1 :
                dp[i][j] = 0
            else:
                dp[i][j] = dp[i-1][j]+dp[i][j-1]
    return dp[-1][-1]
```

```python
grid = [ [1,0,0,0] ,[1,0,1,0] ,[0,0,0,0]]
print( uniquePathWithObstacles( grid ) )
```
