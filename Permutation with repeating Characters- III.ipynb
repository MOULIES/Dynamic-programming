```python
# Permutations With Repeating Characters
```

```python
def permutation_repeat( text , k ):
    res = []
    
    def _permutation_repeat( text , prefix, n , k):
        if k == 0 : # base, len(prefix) == len(text)
            res.append( prefix )
            return
            
        for i in range( n ):
            new_prefix = prefix+ str(text[i]) ## a, aa, aaa, aab, aac ab, aba, abb, abc
            
            # print(new_prefix)
            _permutation_repeat(text, new_prefix, n, k-1)
            
    _permutation_repeat( text , '', len( text ) , k)
    return res
    
permutation_repeat([1,2,3], 3)            
```

```python
from itertools import product

for _set in product(list('12'), repeat=3):
    print(''.join(_set))
```

```python
def product2(*args, repeat=1):
    # product('ABCD', 'xy') --> Ax Ay Bx By Cx Cy Dx Dy
    # product(range(2), repeat=3) --> 000 001 010 011 100 101 110 111
    pools = [ tuple(pool) for pool in args] * repeat
    print(pools, 'hello' , args)
    result = [[]]
    for pool in pools:
        print( result , 'start')
        result = [x+[y] for x in result for y in pool]
        print( result , 'end' )
    for prod in result:
        yield tuple(prod)
        
a= product2( [1,2] , repeat = 3 )     
print( *a )

# for i in a:
#     print( i )
```
