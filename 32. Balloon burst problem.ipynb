```python
def maxCoins(nums):
    nums = [1] + nums + [1]
    
    mem = {}
    
    def dfs(l,r):
        if l> r: return 0 
        
        if (l,r) in mem: return mem[(l,r)]
        
        dp[(l,r)] = 0
        
        for i in range(l,r+1):
            coins = nums[l-1]+ nums[i]+nums[r+1]
            coins += dfs(l,i-1)+dfs(i+1, r)
            dp[(l,r)] = max( coins, dp[(l,r)] )
        return dp[(l,r)]
    
    return dfs(1, len(nums)-2)
```
