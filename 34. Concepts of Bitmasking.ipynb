```python
def bitMasking( s ):
    n = len(s)
    res = []
    for mask in range( 1<<n ):  #2^N
        curr_set= ""
        for j in range(n): #N
            if mask & 1<<j:
                curr_set+= s[j]
        res.append( curr_set )
    return res

#time complexity : O(N. 2^N)
```

```python
s = 'abc'
print( bitMasking( s ) )
```
