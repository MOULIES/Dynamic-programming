```python
def rod_cut_recursive( price , n ):
    length = [ i+1 for i in range(n) ]
    maxlen = n
    dp = [  [ 0]*(n+1)  for _ in range( n+1)]
    
    def ubk( price, length , maxlen , n):
        if n== 0 or maxlen == 0 :
            return 0
        if  maxlen  < length[n-1] :
            dp[n][maxlen] = ubk( price, length , maxlen , n-1 )
        else:
            dp[n][maxlen] = max( ubk( price, length , maxlen , n-1 ) ,  
                                    price[n-1]  + ubk( price, length , maxlen-length[n-1] , n ) )
            
        return dp[n][maxlen] 
    return ubk( price, length , maxlen , n)
```

### Dynamic programming - I

```python
def rod_cutting( price, n ):
    dp = [[0]*(n+1) for i in range(n+1)]
    
    for i in range(1, n+1):
        for j in range( 1, n+1):
            if j < i :
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = max( dp[i-1][j] , price[i-1]+dp[i][ j-i] )
#     for i in dp:
#         print( *i )
    return dp[-1][-1]
```
