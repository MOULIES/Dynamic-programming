```python
def minOperations( nums, x ):
    mem = { }
    import math
    def solve( nums, x ,l , r, count ):
        if (l,r) in mem :
            return mem[(l,r)]
        
        if x == 0 :
            return count
        
        if x< 0 or l> r: #out of bounds
            return math.inf
        
        left = solve( nums, x-nums[l] ,l+1 , r, count+1 ) 
        right = solve( nums, x-nums[r] ,l , r-1 , count+1 ) 
        
        return min( left, right)
    
    ans = solve( nums, x ,0 , len(nums)-1 , 0 )
    return ans if ans != math.inf else -1
```

```python
nums = [1,1,4,2,3]
x  = 5
minOperations( nums, x )
```

```python
nums = [5,3,1,5]
x  = 1
minOperations( nums, x )
```

```python
def minOperations_dp( nums , x):
    n = len( nums )
    cum_sum = 0
    mem = {}
    for i in range( n ):
        cum_sum += nums[i]
        mem[cum_sum ] = i
        
    if x > cum_sum :
        return -1
    
    mem[0] = 0
    longest = 0
    cum_sum -= x 
    val = 0
    
    for i in range(n):
        val += nums[i]
        
        if (val - cum_sum) in mem:
            if (val - cum_sum) == 0:
                longest = max( longest , i-mem[val - cum_sum] + 1 )
            else:
                longest = max( longest , i-mem[val - cum_sum]  )
    return (  n  if cum_sum == 0  else -1  ) if longest == 0  else n-longest
```

```python
nums = [1,1,4,2,3]
x  = 5
minOperations_dp( nums, x )
```

```python
nums = [5,3,1,5]
x  = 1
minOperations_dp( nums, x )
```
