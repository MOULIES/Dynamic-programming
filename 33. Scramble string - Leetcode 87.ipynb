```python
def isScramble(s1: str, s2: str) -> bool:
    n = len(s1)
    if n != len(s2): return False
    
    if s1 == s2 : return True
    
    for i in range(1, n):
        if isScramble( s1[:i] , s2[:i]) and isScramble( s1[i:] , s2[i:]):
            return True
        
        #flip case
        if  isScramble( s1[:i] , s2[n-i:] ) and isScramble( s1[i:] , s2[:n-i]):
            return True
    return False
```

```python
S1 = "coder"
S2 = "ocred"
print( isScramble(S1, S2) )
```
